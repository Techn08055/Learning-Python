### Gaussian Distribution

It is one of the most common distribution in Data Science. It is also one of the common distributions in our daily life. This distribution is a bell shaped curve symmetrical around the mean. The mean and standard deviation forms the major part in this distribution. Due to its real life occurances many machine learning models are created based on this.

Its general form it's pdf is:

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statistics
import seaborn as sns
from numpy import random
```

```python
x = np.arange(-20, 20)
mean = 0  
plt.plot(x, stats.norm.pdf(x, mean, 10))
plt.xlabel('Distribution')
plt.ylabel('Probability')
plt.show()
```

### Uniform Distribution

I a uniform distribution the probability of occurance of an event is same in the interval it is defined. There are few examples like : tossing of a unbiased coin, dice etc.

General form of its pdf is:

image.png

```python
sns.distplot(random.uniform(low=-2, high=1.0, size=10000), hist=False)
plt.xlabel('Distribution')
plt.ylabel('Probability')
plt.show()
```

### Binomial Distribution

It can be defined as the probability of success/failure of an experiment repeated multiple times independently.It has two possible outcomes. The probability depends on the number of times the experiment is repeated and the probability of one of the classes. 

It's general form is:

image.png

```python
n = 6
p = 0.4

r_values = list(range(n + 1))
dist = [stats.binom.pmf(r, n, p) for r in r_values ]
plt.plot(r_values, dist,'-*')
plt.xlabel('Distribution')
plt.ylabel('Probability')
plt.show()
```

Bernoulli Distribution is a special case of Binomial Distribution

### Log Normal Distribution

Lognormal Distribution is a distribution of random variables whose natural logarithm is Normal/Gaussian distribution. Many of the natural phenomena is an example where each small events add to give a major result. It has many applications in understanding human behaviours, biology and medicine


```python
random.lognormal(mean=0.0, sigma=1.0, size=None)
```

```python
x = np.linspace(0, 5, 100) 
dist = stats.lognorm.pdf(x,.5,3 )
plt.plot(x, dist)
plt.xlabel('Distribution')
plt.ylabel('Probability')
plt.show()
```

### Pareto Distribution

80 percent of result is due to 2 percent incidence


```python
x = np.linspace(0, 5, 100) 
dist = stats.pareto.pdf(x,1.5,3 )
plt.plot(x, dist)
plt.xlabel('Distribution')
plt.ylabel('Probability')
plt.show()
```

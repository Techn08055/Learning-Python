### Modules

They are `.py` files containing collection of functions and global variables.
Modules help us to break down the code into smaller reusable chunks. It reduce rewriting of code sice we can import the functionality by just importing these modules.

Importing a Module

You can use the keyword `import` to add the module to your code

```python
import Myname #imported my dummy module Myname(Myname.py)
```

You can access the functions in a module using `.` operator.

```python
Myname.myname("Raj")
```

There are many expert written modules in python whose power can be leverged by you in your own code.

https://docs.python.org/3/py-modindex.html

Let us go through few examples...

```python
import numpy
print(numpy.abs(-7)) #we just called the function abs inside numpy
```

#### Renaming The Modules

You can create an alias for the modules using the keyword `as`. 

```python
import Myname as m
m.myname("Raj")
```

```python
import numpy as np
print(np.abs(-7))
```

Most of the official modules are very large and contains lots of function that might be unnecessary for you at the time. You can import your specific function in module using `from`...`import` statement

```python
from Myname import myname
myname("Raj")
```

```python
from numpy import abs
print(abs(-7))
```

You can also import all the functions in a module using `*` with `from`... `import` statement

```python
from numpy import *
print(abs(-7))
```
